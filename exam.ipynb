`[0.2]` Create a new column `whoWon` which contains `"red"` or `"blue"` based on the `blueWins` column.

````python
def whowonfunc(lol_data):
    if lol_data['blueWins'] == 0:
        val = "red"
    else:
        val = "blue"
    return val



lol_data['whoWon'] = lol_data.apply(whowonfunc, axis=1)
````

`[0.2]` Create a new column `firstKill` which contains `"red"` or `"blue"` based on either `blueFirstBlood` or `redFirstBlood`.

````python
def whofirstkillfunc(lol_data):
    if lol_data['blueFirstBlood'] == 1:
        val = "blue"
    else:
        val = "red"
    return val

lol_data['firstKill'] = lol_data.apply(whofirstkillfunc, axis=1)
````

## Task 5

`[0.8]` Create a new variable `lol_data_group` where you groupby `whoWon` and `firstKill`, with the following aggregations:

- `blueTotalGold` - sum
- `redTotalGold` -  sum
- `blueKills` - mean

````python
lol_data_group = lol_data.groupby(["whoWon","firstKill"]).agg({
    'blueTotalGold': 'sum',
    'redTotalGold': 'sum',
    'blueKills': 'mean',
    'redKills': 'mean',
    'blueGoldPerMin': 'max',
    'redGoldPerMin': 'min'
})
lol_data_group
````

`[0.2]` **Rename** (use good function here) the columns of `lol_data_group` to `blueTotalGoldSum`, `redTotalGoldSum`, `blueKillsMean`, `redKillsMean`, `blueGoldPerMinMax`, `redGoldPerMinMin`.
